#### Creating a dataframe and generate random observations

```python
number_of_objects = 30
number_of_observations = 100
number_of_axes = 2
names_of_axes = ['x', 'y']

number_of_positive_objects = number_of_objects // 2
number_of_negative_objects = number_of_objects // 2

objects = ''
axis_index = []
```

```python
for i in range(1, number_of_objects+1):
    objects += str(i) + ' ' + str(i) + ' '
    for j in names_of_axes:
        axis_index.append(j)
    
objects_index = objects.split()
hier_index = list(zip(objects_index,axis_index))
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
df = pd.DataFrame(np.random.randn(number_of_objects*number_of_axes,number_of_observations), index=hier_index)
df.columns.name = 'Observation'
df.index.names = ['Object', 'Axis']
#df['class_label'] = 1
df.head(10)
```
